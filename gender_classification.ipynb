```python
import pandas as pd
import numpy as np
```

```python
#C:\Users\uqurk
data = pd.read_csv("C:/Users/uqurk/desktop/sheet1.csv")
data.head()
```

```python
df1 = [1 if i == "F" else 0 for i in data["Gender"] ]
df_gender = pd.DataFrame(data = {"Gender": df1})
df2 = pd.get_dummies(data.loc[:,:'Favorite Soft Drink'])
data = pd.concat([df2,df_gender], axis=1)
```

```python
data
```

```python
x = data.drop("Gender", axis=1)
y = data["Gender"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
#Lojistik Regresyon
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
print(lr.score(x_test, y_test))
print(lr.score(x_train, y_train))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
accuracy_test = []
accuracy_train = []
for i in range(1,53):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    accuracy_test.append(knn.score(x_test, y_test))
    accuracy_train.append(knn.score(x_train, y_train))
    
print(max(accuracy_test))
print(accuracy_train[accuracy_test.index(np.max(accuracy_test))])
```

```python
#SVM
from sklearn.svm import SVC
dvm = SVC(C = 100,gamma = 0.01)
dvm.fit(x_train,y_train)
print(dvm.score(x_test, y_test))
print(dvm.score(x_train, y_train))
```
